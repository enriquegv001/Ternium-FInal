```python
# IMPORTAMOS LIBRERIAS

import pandas as pd
import numpy as np
import pickle
```

```python
# MANDAR A LLAMAR MODELO

with open('model.pkl' , 'rb') as f:
    DT = pickle.load(f)
```

Escribe la direccion y el nombre de tu archivo .csv entre las comillas

```python
file = "ejemplo_Ternium.csv"
x = pd.read_csv(file)
x
```

```python
# MANEJAMOS CSV

x = pd.read_csv(file)

ID = x["ID Candidato"]
ID = np.array(ID)

del(x["Unnamed: 0"])
del(x["ID Candidato"])
```

```python
# EMPLEAR MODELO

result = DT.predict(x)

result_df = pd.DataFrame({'ID':ID, 'Ingresado':result})

result_df = result_df.replace({'Ingresado': 1}, "Si")
result_df = result_df.replace({'Ingresado': 0}, "No")

result_si = result_df[result_df["Ingresado"] == "Si"]
result_no = result_df[result_df["Ingresado"] == "No"]

```

```python
# Exportar resultados
result_df.to_csv("Resultado_Ingresados.csv")
result_si.to_csv("Resultado_Si_Ingresados.csv")
result_no.to_csv("Resultado_No_Ingresados.csv")
```
